Apply backpropagation plus gradient desect using sigmoid as an activation   function for (AND GATE).

```python
import numpy as np
import matplotlib.pyplot as plt

# AND gate dataset
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([[0], [0], [0], [1]])

# Sigmoid activation
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Derivative of sigmoid
def sigmoid_derivative(a):
    return a * (1 - a)

# Training function
def train(lr, epochs=1000):
    np.random.seed(0)
    weights = np.random.randn(2, 1)
    bias = np.random.randn(1)
    errors = []

    for epoch in range(epochs):
        # Forward pass
        z = np.dot(X, weights) + bias
        y_pred = sigmoid(z)

        # Error (MSE)
        error = np.mean((y - y_pred) ** 2)
        errors.append(error)

        # Backpropagation
        d_error = y_pred - y
        d_pred = d_error * sigmoid_derivative(y_pred)

        # Gradients
        dW = np.dot(X.T, d_pred) / len(X)
        dB = np.mean(d_pred)

        # Gradient descent update
        weights -= lr * dW
        bias -= lr * dB

    return errors

# Learning rates to compare
learning_rates = [0.1, 0.5, 1.0]

# Plotting
plt.figure(figsize=(8, 5))

for lr in learning_rates:
    errors = train(lr)
    plt.plot(errors, label=f"LR = {lr}")

plt.xlabel("Epochs")
plt.ylabel("Mean Squared Error")
plt.title("Error vs Epochs for AND Gate (Sigmoid + Backprop)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Sigmoid activation function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Derivative of sigmoid
def sigmoid_derivative(y):
    return y * (1 - y)

# Training data for AND gate
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([[0], [0], [0], [1]])

# Initialize weights and bias
np.random.seed(1)
w = np.random.rand(2, 1)   # weights w1, w2
b = np.random.rand(1)      # bias
lr = 0.1                   # learning rate

# Training loop
epochs = 10000
for _ in range(epochs):
    for i in range(len(X)):
        # Forward pass
        z = np.dot(X[i], w) + b
        y_hat = sigmoid(z)

        # Error
        error = y[i] - y_hat

        # Backpropagation
        delta = error * sigmoid_derivative(y_hat)

        # Gradient descent update
        w += lr * delta * X[i].reshape(2, 1)
        b += lr * delta

# Testing
print("Final weights:\n", w)
print("Final bias:\n", b)

print("\nPredictions:")
for i in range(len(X)):
    z = np.dot(X[i], w) + b
    output = sigmoid(z)
    print(f"Input: {X[i]} -> Output: {output[0]:.4f}")

```

Apply backpropagation plus gradient desect using sigmoid as an activation function for (OR GATE).

```python
import numpy as np
import matplotlib.pyplot as plt

# OR Gate Dataset
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# OR gate outputs
y = np.array([0, 1, 1, 1])

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Training function
def train_model(learning_rate, epochs=100):
    w = np.array([0.1, 0.2])
    b = 0.3
    errors = []

    for epoch in range(epochs):
        total_error = 0

        for i in range(len(X)):
            # Forward pass
            z = np.dot(X[i], w) + b
            y_pred = sigmoid(z)

            # Binary Cross Entropy Loss
            loss = -(y[i] * np.log(y_pred + 1e-9) +
                     (1 - y[i]) * np.log(1 - y_pred + 1e-9))
            total_error += loss

            # Backpropagation
            dz = y_pred - y[i]
            dw = dz * X[i]
            db = dz

            # Gradient Descent update
            w -= learning_rate * dw
            b -= learning_rate * db

        errors.append(total_error)

    return errors, w, b


# Learning rate
lr = 1.9
errors, w, b = train_model(lr)

# Plot error curve
plt.plot(errors)
plt.xlabel("Epoch")
plt.ylabel("Total Error")
plt.title(f"Error vs Epoch (OR Gate, LR = {lr})")
plt.show()

# Testing
print("Final Weights:", w)
print("Final Bias:", b)

print("\nPredictions:")
for i in range(len(X)):
    z = np.dot(X[i], w) + b
    print(X[i], "->", sigmoid(z))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# OR gate dataset
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([[0], [1], [1], [1]])

# Sigmoid activation function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Derivative of sigmoid
def sigmoid_derivative(a):
    return a * (1 - a)

# Training function using backpropagation + gradient descent
def train(lr, epochs=1000):
    np.random.seed(0)
    weights = np.random.randn(2, 1)
    bias = np.random.randn(1)
    errors = []

    for epoch in range(epochs):
        # Forward propagation
        z = np.dot(X, weights) + bias
        y_pred = sigmoid(z)

        # Mean Squared Error
        error = np.mean((y - y_pred) ** 2)
        errors.append(error)

        # Backpropagation
        d_error = y_pred - y
        d_pred = d_error * sigmoid_derivative(y_pred)

        # Gradients
        dW = np.dot(X.T, d_pred) / len(X)
        dB = np.mean(d_pred)

        # Gradient descent update
        weights -= lr * dW
        bias -= lr * dB

    return errors

# Different learning rates
learning_rates = [0.1, 0.5, 1.0]

# Plot Error vs Epochs
plt.figure(figsize=(8, 5))

for lr in learning_rates:
    errors = train(lr)
    plt.plot(errors, label=f"LR = {lr}")

plt.xlabel("Epochs")
plt.ylabel("Mean Squared Error")
plt.title("Error vs Epochs for OR Gate (Sigmoid + Backpropagation)")
plt.legend()
plt.grid(True)
plt.show()
```
